### Exploratory Visualization

```python
plt.figure(figsize=(14,60))

sns.boxplot(y='neighbourhood_cleansed', x='number_of_reviews', data=df)

plt.xlabel('Number of Reviews', fontsize=16, fontweight='bold')
plt.ylabel('Neighborhood', fontsize=16, fontweight='bold')
```
